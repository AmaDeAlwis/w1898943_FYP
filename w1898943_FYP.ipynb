```python
#numerical features for IQR calculation
numerical_features = breast_cancer_cleaned.select_dtypes(include=['number']).columns

#quantiles
Q1 = breast_cancer_cleaned[numerical_features].quantile(0.25)
Q3 = breast_cancer_cleaned[numerical_features].quantile(0.75)
IQR = Q3 - Q1

#detecting outliers
outliers_iqr = ((breast_cancer_cleaned[numerical_features] < (Q1 - 1.5 * IQR)) | (breast_cancer_cleaned[numerical_features] > (Q3 + 1.5 * IQR)))

#Remove rows with outliers in any numerical feature
cleaned_data_iqr = breast_cancer_cleaned[~outliers_iqr.any(axis=1)]
```

```python
#removing duplicates
breast_cancer_cleaned = breast_cancer_cleaned.drop_duplicates()
```

```python
#keeps breast cancer in "Cancer type"
breast_cancer_cleaned = breast_cancer_cleaned[breast_cancer_cleaned['Cancer Type'] == 'Breast Cancer']
```

```python
#remove unnecessary columns
drop_columns = ["Patient ID","Cancer Type","Cancer Type Detailed","Cellularity","Pam50 + Claudin-low subtype","Cohort",
"ER Status","Neoplasm Histologic Grade","HER2 Status","Tumor Other Histologic Subtype","Integrative Cluster","Primary Tumor Laterality",
"Mutation Count","Nottingham prognostic index","Oncotree Code","Overall Survival Status","Relapse Free Status (Months)","Relapse Free Status",
"Sex","3-Gene classifier subtype","Tumor Size"]

breast_cancer_cleaned = breast_cancer_cleaned.drop(columns=drop_columns)
print(breast_cancer_cleaned.head())
```

```python
categorical_columns = breast_cancer_cleaned.select_dtypes(include=['object']).columns

#Remove categorical column contains NaN
breast_cancer_cleaned = breast_cancer_cleaned.dropna(subset=categorical_columns, how='any')

#display unique values
for col in categorical_columns:
    print(f"Unique values in {col}: {breast_cancer_cleaned[col].unique()}")
```

```python
from sklearn.preprocessing import LabelEncoder

#Label Encoding for Binary Columns
binary_columns = ['Chemotherapy','ER status measured by IHC', 'Hormone Therapy','Inferred Menopausal State', 'PR Status', 'Radio Therapy','Patient\'s Vital Status']

label_encoder = LabelEncoder()

#Loop
for col in binary_columns:
    breast_cancer_cleaned[col] = label_encoder.fit_transform(breast_cancer_cleaned[col])

#One Hot Encoding
categorical_columns = ['Type of Breast Surgery', 'HER2 status measured by SNP6']

breast_cancer_cleaned = pd.get_dummies(breast_cancer_cleaned, columns=categorical_columns)
print(breast_cancer_cleaned.head())
```

Graph Neural Network

```python
import torch
from sklearn.metrics.pairwise import euclidean_distances

adj_matrix = euclidean_distances(X_train)

#threshold for distance to decide whether an edge should exist
threshold = np.percentile(adj_matrix, 90)
adj_matrix[adj_matrix > threshold] = 1
adj_matrix[adj_matrix <= threshold] = 0

#Convert adjacency matrix to a PyTorch tensor
adj_matrix = torch.tensor(adj_matrix, dtype=torch.float)

#PyTorch Geometric uses an edge list (edge_index) to represent the graph
edge_index = torch.nonzero(adj_matrix, as_tuple=False).t()

#Convert X_train node feature matrix
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float)

#Create PyTorch Geometric Data object
from torch_geometric.data import Data

graph_data = Data(x=X_train_tensor, edge_index=edge_index)
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data

#Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train.values)

#encoding
label_encoder = LabelEncoder()
y_event_encoded = label_encoder.fit_transform(y_train["Patient's Vital Status"])

#Convert X_train and y_train to PyTorch tensors
X_tensor = torch.tensor(X_scaled, dtype=torch.float)
y_time_tensor = torch.tensor(y_train['Overall Survival (Months)'].values, dtype=torch.float)
y_event_tensor = torch.tensor(y_event_encoded, dtype=torch.float)
```

```python
#Create adjacency matrix using Euclidean distance between patients
adj_matrix = euclidean_distances(X_scaled)
threshold = np.percentile(adj_matrix, 90)
adj_matrix[adj_matrix > threshold] = 1
adj_matrix[adj_matrix <= threshold] = 0
adj_matrix = torch.tensor(adj_matrix, dtype=torch.float)

#Create edge index from adjacency matrix
edge_index = torch.nonzero(adj_matrix, as_tuple=False).t()

#PyTorch Geometric Data object
graph_data = Data(x=X_tensor, edge_index=edge_index)

#GNN model with dual survival time and event status
class SurvivalGNN(nn.Module):
    def __init__(self, in_channels, out_channels_time, out_channels_event):
        super(SurvivalGNN, self).__init__()
        self.conv1 = GCNConv(in_channels, 64)
        self.conv2 = GCNConv(64, 64)
        self.fc_time = nn.Linear(64, out_channels_time)
        self.fc_event = nn.Linear(64, out_channels_event)

    def forward(self, x, edge_index):
        #Apply Graph Convolution layers with ReLU activations
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))

        #Separate output for survival time and event status
        survival_time_output = self.fc_time(x)
        event_status_output = self.fc_event(x)

        return survival_time_output, event_status_output
```
