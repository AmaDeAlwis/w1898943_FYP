```python
#Identify numerical features
numerical_features = breast_cancer_cleaned.select_dtypes(include=['number']).columns

#Calculate IQR
Q1 = breast_cancer_cleaned[numerical_features].quantile(0.25)
Q3 = breast_cancer_cleaned[numerical_features].quantile(0.75)
IQR = Q3 - Q1

#Detect outliers
outliers_iqr = ((breast_cancer_cleaned[numerical_features] < (Q1 - 1.5 * IQR)) |
                (breast_cancer_cleaned[numerical_features] > (Q3 + 1.5 * IQR)))

#Remove rows with outliers in any numerical column
cleaned_data = breast_cancer_cleaned[~outliers_iqr.any(axis=1)]

#how many rows were removed
outliers_removed = after_duplicates_rows - cleaned_data.shape[0]
print(f"Outliers removed: {outliers_removed}")
```

```python
from sksurv.metrics import concordance_index_censored
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sksurv.ensemble import RandomSurvivalForest

X = breast_cancer_cleaned.drop(columns=['Overall Survival (Months)', "Patient's Vital Status"])

#time and event
y_time = breast_cancer_cleaned["Overall Survival (Months)"]
y_event = breast_cancer_cleaned["Patient's Vital Status"]

#Flip the event indicator
y_event_surv = 1 - y_event

#Create a structured array
y_surv = np.array(
    list(zip(y_event_surv.astype(bool), y_time)),
    dtype=[("event", bool), ("time", float)]
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_surv, test_size=0.3, random_state=42
)

#Define the custom scorer
def c_index_score(y_true, y_pred):
    return concordance_index_censored(y_true["event"], y_true["time"], y_pred)[0]

scorer = make_scorer(c_index_score, greater_is_better=True)
```

```python
#tuned model
best_rsf = grid_search.best_estimator_

#Predict survival functions
surv_fns = best_rsf.predict_survival_function(X_test)

#Define target time points
target_times = [60, 120]
surv_probs_at_60 = []
surv_probs_at_120 = []

#Loop through each patient's survival function
for fn in surv_fns:
    prob_60 = np.interp(60, fn.x, fn.y)
    prob_120 = np.interp(120, fn.x, fn.y)
    surv_probs_at_60.append(prob_60)
    surv_probs_at_120.append(prob_120)

#patients' probabilities
for i in range(5):
    print(f"Patient {i+1} -> 5-year: {surv_probs_at_60[i]:.4f}, 10-year: {surv_probs_at_120[i]:.4f}")
```

```python
#Create features and target
X = breast_cancer_cleaned.drop(columns=["Overall Survival (Months)", "Patient's Vital Status"])
y = breast_cancer_cleaned[["Overall Survival (Months)", "Patient's Vital Status"]]

#Train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from lifelines import AalenAdditiveFitter
from sklearn.model_selection import KFold
import warnings

warnings.filterwarnings("ignore")

#Hyperparameter grid
coef_penalizers = [0.001, 0.01, 0.1, 1.0]
smoothing_penalizers = [0.001, 0.01, 0.1, 1.0]

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

#Loop through combinations
for coef_pen in coef_penalizers:
    for smooth_pen in smoothing_penalizers:
        c_indexes = []

        for train_idx, val_idx in kf.split(X_train):

            X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
            y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

            # Combine into training and validation
            train_df = X_tr.copy()
            train_df["Overall Survival (Months)"] = y_tr["Overall Survival (Months)"]
            train_df["event"] = 1 - y_tr["Patient's Vital Status"]

            val_df = X_val.copy()
            val_df["Overall Survival (Months)"] = y_val["Overall Survival (Months)"]
            val_df["event"] = 1 - y_val["Patient's Vital Status"]

            #Train model
            aaf = AalenAdditiveFitter(coef_penalizer=coef_pen, smoothing_penalizer=smooth_pen)
            try:
                aaf.fit(train_df, duration_col="Overall Survival (Months)", event_col="event")
                c_index = aaf.concordance_index_
                c_indexes.append(c_index)
            except:
                continue

        mean_c_index = np.mean(c_indexes) if c_indexes else np.nan
        cv_results.append({
            "coef_penalizer": coef_pen,
            "smoothing_penalizer": smooth_pen,
            "c_index": mean_c_index
        })

#Best hyperparameters
cv_df = pd.DataFrame(cv_results)
best_params = cv_df.loc[cv_df['c_index'].idxmax()]
print("\n Best Hyperparameters:")
print(best_params)
```

```python
from lifelines import AalenAdditiveFitter

#Extract best hyperparameters
best_coef = best_params["coef_penalizer"]
best_smooth = best_params["smoothing_penalizer"]

#Prepare final training data
final_train_data = X_train.copy()
final_train_data["Overall Survival (Months)"] = y_train["Overall Survival (Months)"]
final_train_data["event"] = 1 - y_train["Patient's Vital Status"]

#Train the final model
aah_model = AalenAdditiveFitter(coef_penalizer=best_coef, smoothing_penalizer=best_smooth)
aah_model.fit(final_train_data, duration_col="Overall Survival (Months)", event_col="event")
```

```python
#Predict cumulative hazard
cum_hazards = []
for i in range(len(X_test)):
    row = X_test.iloc[[i]]
    ch = aah_model.predict_cumulative_hazard(row)
    #Use final value (
    cum_hazards.append(ch.values[-1, 0])

cum_hazards = np.array(cum_hazards)

#Flip Patient's Vital Status
event_test = 1 - y_test["Patient's Vital Status"]
time_test = y_test["Overall Survival (Months)"]

#Compute C index on the test set
c_index_test = concordance_index_censored(
    event_test.astype(bool),
    time_test,
    cum_hazards
)[0]

print(f" Test Set C-Index for Aalen's Additive Model: {c_index_test:.4f}")
```

```python
#Convert y_train and y_test
y_train_struct = Surv.from_arrays(
    event=1 - y_train["Patient's Vital Status"],
    time=y_train["Overall Survival (Months)"]
)

y_test_struct = Surv.from_arrays(
    event=1 - y_test["Patient's Vital Status"],
    time=y_test["Overall Survival (Months)"]
)

#Define time grid
epsilon = 1e-5
time_grid = np.linspace(
    y_test["Overall Survival (Months)"].min(),
    y_test["Overall Survival (Months)"].max() - epsilon,
    100
)

#convert to survival probability
all_surv_probs = []

for i in range(len(X_test)):
    row = X_test.iloc[[i]]
    cum_hazard = aah_model.predict_cumulative_hazard(row)
    ch_interp = np.interp(time_grid, cum_hazard.index.values, cum_hazard.values[:, 0])
    surv_probs = np.exp(-np.clip(ch_interp, 0, None))
    all_surv_probs.append(surv_probs)

#Survival probability matrix
pred_surv_probs = np.vstack(all_surv_probs)
```

```python
!pip install -q scikit-survival torch-geometric scikit-optimize

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics.pairwise import euclidean_distances
from skopt import gp_minimize
from skopt.space import Real, Integer
#random seeds
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

#Ensure deterministic behaviour
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

#Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.values)

#Encode event status
label_encoder = LabelEncoder()
y_event_encoded = label_encoder.fit_transform(y["Patient's Vital Status"])

X_train_part, X_test_part, y_train_part, y_test_part = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

#Create tensors
X_train_tensor = torch.tensor(X_train_part, dtype=torch.float)
X_test_tensor = torch.tensor(X_test_part, dtype=torch.float)

#Build graph structures
adj_matrix_train = euclidean_distances(X_train_part)
np.fill_diagonal(adj_matrix_train, 0)
threshold_train = np.percentile(adj_matrix_train, 90)
adj_matrix_train = (adj_matrix_train > threshold_train).astype(int)
adj_matrix_train = torch.tensor(adj_matrix_train, dtype=torch.float)
edge_index_train = torch.nonzero(adj_matrix_train, as_tuple=False).t().contiguous().long()

adj_matrix_test = euclidean_distances(X_test_part)
np.fill_diagonal(adj_matrix_test, 0)
threshold_test = np.percentile(adj_matrix_test, 90)
adj_matrix_test = (adj_matrix_test > threshold_test).astype(int)
adj_matrix_test = torch.tensor(adj_matrix_test, dtype=torch.float)
edge_index_test = torch.nonzero(adj_matrix_test, as_tuple=False).t().contiguous().long()
```

```python
#Create data objects
graph_data_train = Data(x=X_train_tensor, edge_index=edge_index_train)
graph_data_test = Data(x=X_test_tensor, edge_index=edge_index_test)

#GCN model
class SurvivalGNN(nn.Module):
    def __init__(self, in_channels, hidden_channels, dropout_rate=0.3):
        super(SurvivalGNN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc_5yr = nn.Linear(hidden_channels, 1)
        self.fc_10yr = nn.Linear(hidden_channels, 1)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = self.dropout(F.relu(self.conv2(x, edge_index)))
        surv_5yr = torch.sigmoid(self.fc_5yr(x))
        surv_10yr = torch.sigmoid(self.fc_10yr(x))
        return surv_5yr, surv_10yr
```

```python
#loss function
def survival_loss(pred_5yr, pred_10yr, true_5yr, true_10yr):
    bce = nn.BCELoss()
    loss_5yr = bce(pred_5yr.squeeze(), true_5yr.float())
    loss_10yr = bce(pred_10yr.squeeze(), true_10yr.float())
    return (loss_5yr + loss_10yr) / 2

#objective function for Bayesian Optimization
def objective(params):
    hidden_channels, dropout_rate, learning_rate = params
    model = SurvivalGNN(in_channels=X_train_tensor.shape[1], hidden_channels=hidden_channels, dropout_rate=dropout_rate)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    model.train()
    optimizer.zero_grad()

    pred_5yr, pred_10yr = model(graph_data_train)
    loss = survival_loss(pred_5yr, pred_10yr, y_train_5yr, y_train_10yr)

    loss.backward()
    optimizer.step()
    return loss.item()
```

```python
#Prepare survival labels for 5yr and 10yr prediction
y_train_time = y_train_part['Overall Survival (Months)'].values
y_train_event = y_train_part["Patient's Vital Status"].values

#Flip event
y_train_event = 1 - y_train_event

#Create survival labels considering death and censoring
y_train_5yr = torch.tensor(((y_train_time > 60) | (y_train_event == 0)).astype(int), dtype=torch.float)
y_train_10yr = torch.tensor(((y_train_time > 120) | (y_train_event == 0)).astype(int), dtype=torch.float)


#Bayesian Optimization
space = [
    Integer(32, 256, name='hidden_channels'),
    Real(0.1, 0.4, name='dropout_rate'),
    Real(0.0005, 0.01, name='learning_rate')
]

result = gp_minimize(objective, space, n_calls=20, random_state=42)
best_hidden_channels, best_dropout_rate, best_learning_rate = result.x

print("\n Best Hyperparameters:")
print(f" Hidden Channels: {best_hidden_channels}")
print(f" Dropout Rate: {best_dropout_rate}")
print(f" Learning Rate: {best_learning_rate}")
```

```python
#Train the final model
final_model = SurvivalGNN(in_channels=X_train_tensor.shape[1], hidden_channels=best_hidden_channels, dropout_rate=best_dropout_rate)
optimizer = optim.Adam(final_model.parameters(), lr=best_learning_rate)

num_epochs = 150
for epoch in range(num_epochs):
    final_model.train()
    optimizer.zero_grad()

    pred_5yr, pred_10yr = final_model(graph_data_train)
    loss = survival_loss(pred_5yr, pred_10yr, y_train_5yr, y_train_10yr)

    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Train Loss = {loss.item():.4f}")

#Save the final model
torch.save(final_model.state_dict(), "gcn_model.pt")
print("\n Final GCN model saved as gcn_model.pt!")
```

```python
from lifelines import CoxPHFitter

np.random.seed(42)
np.set_printoptions(precision=8, suppress=True)

#Prepare X and y
X = breast_cancer_cleaned.drop(columns=["Overall Survival (Months)", "Patient's Vital Status"])
y = breast_cancer_cleaned[["Overall Survival (Months)", "Patient's Vital Status"]]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

#Scale features
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X.columns, index=X_test.index)
joblib.dump(scaler, 'scaler.pkl')
print("Scaler saved  as scaler.pkl")

#Create training dataframe
train_data = X_train_scaled.copy()
train_data["time"] = y_train["Overall Survival (Months)"]
train_data["event"] = 1 - y_train["Patient's Vital Status"]
```

```python
from lifelines.utils import k_fold_cross_validation
penalizer_values = [0.01, 0.1, 1.0, 10.0, 100.0]
cv_scores = []

for p in penalizer_values:
    cph = CoxPHFitter(penalizer=p)
    scores = k_fold_cross_validation(cph, train_data,
                                     duration_col="time", event_col="event",
                                     k=5, scoring_method="concordance_index")
    mean_score = np.mean(scores)
    cv_scores.append((p, mean_score))
    print(f"Penalizer={p} âžœ Mean C-index: {mean_score:.4f}")

best_penalizer, best_score = max(cv_scores, key=lambda x: x[1])
print(f"\n Best Penalizer: {best_penalizer} with Mean C-index: {best_score:.4f}")
```

```python
cph = CoxPHFitter(penalizer=best_penalizer)
cph.fit(train_data, duration_col="time", event_col="event")
```
